```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("/content/drive/MyDrive/Breast_cancer_data.csv")
df.head()
```

```python
plt.scatter(df.mean_radius,df.diagnosis,marker='+',color='red')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df[["mean_radius"]],df.diagnosis,train_size=0.8)
x_test
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
x_test
```

```python
y_predicted=model.predict(x_test)
model.predict_proba(x_test)
```

```python
model.score(x_test,y_test)
```
